```python
import json
import torch
import numpy as np
import faiss
from tqdm import tqdm
from transformers import BertTokenizer, BertModel

# add our json file paths
JSON_FILES = [
    "sample.json"
]

# out
EMBEDDING_FILE = "stackoverflow_bert_embeddings.npy"
METADATA_FILE = "stackoverflow_bert_metadata.json"


MODEL_NAME = "bert-base-uncased"
BATCH_SIZE = 8  # increase it to occupy the gpu for faster inference
MAX_TOKENS = 512  # max tokens
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"


tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)
model = BertModel.from_pretrained(MODEL_NAME).to(DEVICE)
model.eval()


def load_questions_from_json(files):
    all_questions, metadata = [], []

    for file in tqdm(files, desc="reading JSON files"):
        with open(file, "r", encoding="utf-8") as f:
            data = json.load(f)
            for item in data:
                title = item.get("title", "").strip()
                body = item.get("body", "").strip()
                answers = item.get("answers", "")
                text = f"{title} {body}"
                all_questions.append(text)
                metadata.append({"title": title, "body": body, "answers": answers, "link": item.get("link", ""), "tags": item.get("tags", [])})

    return all_questions, metadata


questions, metadata = load_questions_from_json(JSON_FILES)


def generate_embeddings(text_list, batch_size):
    """Generates embeddings using PyTorch's transformers on GPU with a configurable batch size."""
    all_embeddings = []

    with torch.no_grad():
        for i in tqdm(range(0, len(text_list), batch_size), desc="Generating BERT Embeddings on GPU"):
            batch_texts = text_list[i:i+batch_size]

            inputs = tokenizer(batch_texts, padding=True, truncation=True, max_length=MAX_TOKENS, return_tensors="pt").to(DEVICE)


            outputs = model(**inputs)

            # use [CLS] token representation
            embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()

            all_embeddings.append(embeddings)

    return np.vstack(all_embeddings)

# generate embeddings (on GPU)
embeddings = generate_embeddings(questions, BATCH_SIZE)

# save embeddings and metadata
np.save(EMBEDDING_FILE, embeddings)
with open(METADATA_FILE, "w", encoding="utf-8") as f:
    json.dump(metadata, f, indent=4)

print(f"✅ Saved {embeddings.shape[0]} embeddings to {EMBEDDING_FILE}")
```
